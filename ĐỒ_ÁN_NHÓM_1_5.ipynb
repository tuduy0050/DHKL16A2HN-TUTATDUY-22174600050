```r
pca_result <- prcomp(data_1, scale = TRUE)
print(pca_result)
```

```r
corr_matrix_1 <- cor(data_1)
ggcorrplot(corr_matrix_1)
```

**Áp Dụng PCA**

```r
data_1.pca <- princomp(corr_matrix_1)
summary(data_1.pca)
```

**Lấy Dữ liệu**

```r
data_1.pca$loadings[, 1:2]
```

**Trực quan hóa các thành phần chính**

```r
fviz_eig(data_1.pca, addlabels = TRUE)
```

**Biplot của các thuộc tính**

```r
fviz_pca_var(data_1.pca, col.var = "black")
```

**Đóng góp của mỗi biến**

```r
fviz_cos2(data_1.pca, choice = "var", axes = 1:2)
```

**Biplot kết hợp với cos2**

```r
fviz_pca_var(data_1.pca, col.var = "cos2",
             gradient.cols = c("black", "orange", "green"),repel = TRUE)
```
